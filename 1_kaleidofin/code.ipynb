```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Exercise Data.xlsx', header = 1)
```

```python
df 
```

```python
'''
Total amount recieved by the each months bidder 
We can get this by adding remaining amount that is to 
be distributed among all 25 members to the amount that has been won
'''
net_return = np.array(df['Net amount recd by Bid winner'])
amount_returned_to_everyone = np.array(df['Amount returned to everyone in the group'])
for i in amount_returned_to_everyone:
     net_return += i

df['Total amount returned to the winner'] = net_return
df
```

```python

'''
Since, we are given 25 months we have to 
calculate average amount that is recieved per 
month 
'''

amount_received_per_month = df['Total amount returned to the winner']/25

'''
Multiply it with 12 to calculate the net return of a year.
''' 

df['Annualized Return'] = amount_received_per_month*12
df
```

```python
'''
%Return = (Total amount returned to the winner / Total contribution) * 100
'''

df['Return Percentage for bid winner'] = (df['Total amount returned to the winner']/50000)*100
```
