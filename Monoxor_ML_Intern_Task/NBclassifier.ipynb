```python
import pandas as pd
```

```python
df = pd.read_csv('isSafe.csv')
```

```python
df
```

```python
df['isSafe_num'] = df['isSafe'].apply(lambda x : 1 if x == True else 0)
df['isSafe_num']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(df['req/body/note/desc'], df['isSafe_num'], test_size = 0.1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train, y_train)
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
X_test_count = v.transform(X_test)
```

```python
model.score(X_test_count, y_test)
```
